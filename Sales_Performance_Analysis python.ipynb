# 1. Identify High-Revenue and Low-Revenue Products
 # - Find the best-selling and least-selling products based on total revenue

```python
product_sales = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False)
print("Top 10 Best-Selling Products:")
print(product_sales.head(10))

print("\nBottom 10 Least-Selling Products:")
print(product_sales.tail(10))
```

#Graphically Representation of the Problem Statement

```python
product_sales = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
product_sales.head(10).plot(kind='bar', color='#5d8aa8')
plt.title("Top 10 Best-Selling Products")
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
product_sales.tail(10).plot(kind='bar', color='#007aa5')
plt.title("Bottom 10 Least-Selling Products")
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)
plt.show()
```

 # 2. *Determine Which Product Categories Drive the Most Profit*  
   # - Analyze total profit across different product categories.  

```python
category_profit = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
print("\nTotal Profit by Category:")
print(category_profit)
```

# Graphical Representation of 2nd Problem Statement

```python
plt.figure(figsize=(8, 6))
category_profit.plot(kind='bar', color='#007aa5')
plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)
plt.show()
```

# 3. *Analyze the Effect of Discounts on Sales and Profitability*  
  # - Determine if higher discounts lead to higher sales or lower profits.  

```python
sns.scatterplot(data=df, x="Discount", y="Sales")
plt.title("Discount vs Sales")
plt.show()

sns.scatterplot(data=df, x="Discount", y="Profit")
plt.title("Discount vs Profit")
plt.show()
```

# 4. *Find the Most Valuable Customers Based on Total Sales*  
   # - Identify customers who contribute the most to total revenue.  

```python
customer_sales = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False)
print("\nTop 10 Most Valuable Customers:")
print(customer_sales.head(10))
```

```python
plt.figure(figsize=(12, 6))
customer_sales.head(10).plot(kind='bar', color='#007aa5')
plt.title("Top 10 Most Valuable Customers")
plt.xlabel("Customer Name")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)
plt.show()
```

# 5. *Analyze Customer Behavior Across Different Segments*  
  # - Compare total sales and profit for customer segments (Consumer, Corporate, Home Office).

```python
segment_sales_profit = df.groupby("Segment")[["Sales", "Profit"]].sum()
print("\nSales and Profit by Customer Segment:")
print(segment_sales_profit)
```

```python
segment_sales_profit.plot(kind='bar', figsize=(8, 6))
plt.title("Sales and Profit by Customer Segment")
plt.xlabel("Segment")
plt.ylabel("Amount")
plt.xticks(rotation=0)
plt.show()
```

# 6. *Repeat Customers Analysis*  
 #  - Identify customers who made multiple purchases and their total contribution to sales.

```python
repeat_customers = df["Customer Name"].value_counts()
print("\nCustomers with Multiple Purchases:")
print(repeat_customers[repeat_customers > 1])
```

# Top 10 Repeat Customers graphical representation

```python
plt.figure(figsize=(12, 6))
repeat_customers[repeat_customers > 1].head(10).plot(kind='bar', color='#007aa5')
plt.title("Top 10 Repeat Customers")
plt.xlabel("Customer Name")
plt.ylabel("Number of Purchases")
plt.xticks(rotation=90)
plt.show()
```

# 7. *Compare Sales Across Different States and Regions*  
  #  - Identify which regions generate the most and least revenue.  

```python
state_sales = df.groupby("State")["Sales"].sum().sort_values(ascending=False)
print("\nSales by State:")
print(state_sales)

region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
print("\nSales by Region:")
print(region_sales)
```

# Top 10 States by Sales

```python
plt.figure(figsize=(12, 6))
df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10).plot(kind='bar', color='#007aa5')
plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)
plt.show()
```

# Sales by Region

```python
plt.figure(figsize=(8, 6))
df.groupby("Region")["Sales"].sum().sort_values(ascending=False).plot(kind='bar', color='#007aa5')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.show()
```

# 8. *Find the Most Profitable Cities*  
  # - Rank cities based on their profit contribution.  

```python
city_profit = df.groupby("City")["Profit"].sum().sort_values(ascending=False)
print("\nMost Profitable Cities:")
print(city_profit.head(10))
```

```python
plt.figure(figsize=(12, 6))
df.groupby("City")["Profit"].sum().sort_values(ascending=False).head(10).plot(kind='bar', color='#007aa5')
plt.title("Top 10 Most Profitable Cities")
plt.xlabel("City")
plt.ylabel("Total Profit")
plt.xticks(rotation=90)
plt.show()
```

# 9. *Analyze Monthly and Yearly Sales Trends*  
  # - Identify seasonality in sales over months and years.  

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], format = "%d-%m-%Y")
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month

yearly_sales = df.groupby("Year")["Sales"].sum()
print("\nYearly Sales Trend:")
print(yearly_sales)

monthly_sales = df.groupby("Month")["Sales"].sum()
print("\nMonthly Sales Trend:")
print(monthly_sales)
```

```python
yearly_sales.plot(kind='line', marker='o', figsize=(10, 6))
plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.grid()
plt.show()

monthly_sales.plot(kind='line', marker='o', figsize=(10, 6))
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

# 10. *Identify Seasonal Demand Fluctuations*  
  # - Detect months or periods with high or low demand.  

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

# 11. *Shipping Performance Analysis*  
  # - Compare the delivery time across different shipping modes

```python
shipping_performance = df.groupby("Ship Mode")["Ship Date"].count()
print("\nShipping Performance by Mode:")
print(shipping_performance)
```

```python
df.groupby("Ship Mode")["Ship Date"].count().plot(kind='bar', color='#007aa5', figsize=(8, 6))
plt.title("Shipping Performance by Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

# 12. *Profitability of Large Orders vs. Small Orders*  
  #  - Compare the profit margins of bulk orders vs. single-item orders.

```python
df["Order Size"] = df["Quantity"]
order_profit = df.groupby("Order Size")["Profit"].mean()
print("\nProfitability by Order Size:")
print(order_profit)
```

```python
df.groupby("Quantity")["Profit"].mean().plot(kind='line', marker='o', figsize=(10, 6), color='#007aa5')
plt.title("Profitability by Order Size")
plt.xlabel("Quantity")
plt.ylabel("Average Profit")
plt.grid()
plt.show()
```
